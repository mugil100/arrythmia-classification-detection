```python
##before removing classess

import os
import glob
from collections import defaultdict

# Define dataset folder (CHANGE this to your dataset location)
dataset_folder = "ecg_file"

# Get all .hea files in the dataset
hea_files = glob.glob(os.path.join(dataset_folder, "*.hea"))
hea_files = hea_files[1:20001]
# Dictionary to store class counts
class_counts = defaultdict(int)

# Function to extract SNOMED CT codes from a .hea file
def get_snomed_codes(hea_file):
    with open(hea_file, "r") as f:
        for line in f:
            if line.startswith("#Dx:"):  # The line that contains the diagnosis codes
                snomed_codes = line.strip().split(":")[1].split(",")
                return [code.strip() for code in snomed_codes]  # Clean spaces
    return []

# Iterate over all .hea files and count occurrences of each class
for hea_file in hea_files:
    snomed_codes = get_snomed_codes(hea_file)
    for code in snomed_codes:
        class_counts[code] += 1  # Increment count for each class

# Print class-wise sample count
print("Class-wise Sample Count:")
for snomed_code, count in sorted(class_counts.items(), key=lambda x: x[1], reverse=True):
    print(f"SNOMED CT Code {snomed_code}: {count} samples")
```

```python
import shutil
import os

def copy_files(source_dir, destination_dir, file_list):
    """
    Copies a list of files from source_dir to destination_dir.

    :param source_dir: Path to the source directory
    :param destination_dir: Path to the destination directory
    :param file_list: List of filenames to copy
    """
    # Ensure the destination directory exists
    os.makedirs(destination_dir, exist_ok=True)

    for file_name in file_list:
        file = file_name.split('\\')[-1]
        file_mat = file.replace('hea','mat')
        source_path = os.path.join(source_dir, file_mat)
        destination_path = os.path.join(destination_dir, file_mat)

        # Check if the file exists before copying
        if os.path.exists(source_path):
            shutil.copy2(source_path, destination_path)
            print(f"Copied: {file_name}")
        else:
            print(f"File not found: {file_name}")

# Example usage
source_directory = "ecg_file"
destination_directory = "ecg_2"

files_to_copy = hea_files[:19980] # Modify with actual file names

copy_files(source_directory, destination_directory, files_to_copy)
```
