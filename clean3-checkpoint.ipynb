```python
#removal of classess

import os
import glob
from collections import defaultdict

# Define dataset folder (CHANGE this to your dataset location)
dataset_folder = r"D:\denoised_data"

# Get all .hea files in the dataset
hea_files = glob.glob(os.path.join(dataset_folder, "*.hea"))
hea_files = hea_files[0:20001]
# Dictionary to store class counts
class_counts = defaultdict(int)

# Function to extract SNOMED CT codes from a .hea file
def get_snomed_codes(hea_file):
    with open(hea_file, "r") as f:
        for line in f:
            if line.startswith("#Dx:"):  # The line that contains the diagnosis codes
                snomed_codes = line.strip().split(":")[1].split(",")
                return [code.strip() for code in snomed_codes]  # Clean spaces
    return []

# First pass: count occurrences of each class
for hea_file in hea_files:
    snomed_codes = get_snomed_codes(hea_file)
    for code in snomed_codes:
        class_counts[code] += 1  # Increment count for each class

# Filter classes with less than 20 samples
filtered_classes = {code for code, count in class_counts.items() if count >= 20}

# Process files again to remove low-count classes
for hea_file in hea_files:
    snomed_codes = get_snomed_codes(hea_file)
    retained_codes = [code for code in snomed_codes if code in filtered_classes]
    
    if not retained_codes:
        os.remove(hea_file)  # Remove the file if no valid classes remain
    else:
        # Rewrite the .hea file with updated class list
        with open(hea_file, "r") as f:
            lines = f.readlines()
        
        with open(hea_file, "w") as f:
            for line in lines:
                if line.startswith("#Dx:"):
                    f.write(f"#Dx: {', '.join(retained_codes)}\n")
                else:
                    f.write(line)

# Print class-wise sample count after filtering
print("Filtered Class-wise Sample Count:")
for snomed_code, count in sorted(class_counts.items(), key=lambda x: x[1], reverse=True):
    if snomed_code in filtered_classes:
        print(f"SNOMED CT Code {snomed_code}: {count} samples")
```
