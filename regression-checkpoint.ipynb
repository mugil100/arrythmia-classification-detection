```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_moons
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv1D, MaxPooling1D, Flatten, Dense, BatchNormalization, Dropout
from sklearn.model_selection import train_test_split
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import confusion_matrix,roc_curve,auc
from sklearn.svm import SVC
```

```python
xb,yb = make_blobs(n_samples = 1000,centers = 2,cluster_std = 3.6,random_state = 42)
plt.scatter(xb[:,0],xb[:,1],c=yb)
```

```python
xm,ym = make_moons(n_samples = 1500,noise = 0.3,random_state = 42)
plt.scatter(xm[:,0],xm[:,1],c=ym)
```

```python
x1 = np.linspace(0,10,600)
m = 3

x2 = [i*m + 4 + np.random.uniform(-3,3) for i in x1]
x3 = [i*m + 8 + np.random.uniform(-3,3) for i in x1]

y1 = [0 for i in x1]
y2 = [1 for i in x1]

yl = np.array(y1 + y2)

x4 = 2*list(x1)
x5 = x2 + x3
xl = np.array([[x4[i],x5[i]] for i in range(len(x4))])
```

```python
plt.scatter(xl[:,0],xl[:,1],c=yl)
```

```python
xm_train,xm_test,ym_train,ym_test = train_test_split(xm,ym,test_size=0.3,random_state=42)
xl_train,xl_test,yl_train,yl_test = train_test_split(xl,yl,test_size=0.3,random_state=42)
xb_train,xb_test,yb_train,yb_test = train_test_split(xb,yb,test_size=0.3,random_state=42)
```

```python
def dnn(input_shape,activation_hidden):
    input_layer = Input(shape = input_shape)
    x = Flatten()(input_layer)
    x = Dense(64,activation = activation_hidden)(x)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)
    x = Flatten()(input_layer)
    x = Dense(32,activation = activation_hidden)(x)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)
    output_layer = Dense(1,activation = None)(x)
    
    model = Model(input_layer,output_layer)
    
    model.compile(optimizer = Adam(0.01),loss='mse',metrics=['accuracy','mae'])
    return model
    
```

```python
bs = (xb.shape[1],)
ls = (xl.shape[1],)
ms = (xm.shape[1],)
```

```python
def report(type,x_test,y_test,y_pred):
    y_pred1 = [1 if i >= 0.5 else 0 for i in y_pred]
    tp,fn,fp,tn = 0,0,0,0

    for i in range(len(y_test)):
        if y_pred1[i] == 1 and y_test[i] == 1:
            tp += 1 
        if y_pred1[i] == 0 and y_test[i] == 1:
            fn +=1
        if y_pred1[i] == 0 and y_test[i] == 0:
            tn += 1
        if y_pred1[i] == 1 and y_test[i] == 0:
            fp += 1

    print(type,":")
    print(f"tp = {tp} , fn = {fn}")
    print(f"fp = {fp} , tn = {tn}")
    
    
    plt.figure(figsize=(20,6))
    plt.subplot(1,2,1)
    plt.scatter(x_test[:,0],x_test[:,1],c=y_test)
    plt.title('true')
    
    plt.subplot(1,2,2)
    plt.scatter(x_test[:,0],x_test[:,1],c=y_pred1)
    plt.title('predicted')
    plt.suptitle(type)
    plt.show()
    
```

```python
ml1 = dnn(ls,None)
ml1.fit(xl_train,yl_train,epochs = 20, validation_split = 0.2)

ml2 = dnn(ls,'relu')
ml2.fit(xl_train,yl_train,epochs = 20, validation_split = 0.2)
```

```python
yl_pred1 = ml1.predict(xl_test)
yl_pred2 = ml2.predict(xl_test)
```

```python
report("linear data linear regression",xl_test,yl_test,yl_pred1)
report("linear data non-linear regression",xl_test,yl_test,yl_pred2)
```

```python
mm1 = dnn(ms,None)
mm1.fit(xm_train,ym_train,epochs = 20, validation_split = 0.2)

mm2 = dnn(ms,'relu')
mm2.fit(xm_train,ym_train,epochs = 20, validation_split = 0.2)
```

```python
ym_pred1 = mm1.predict(xm_test)
ym_pred2 = mm2.predict(xm_test)
```

```python
report('moons linear regression',xm_test,ym_test,ym_pred1)
print("")
report('moons non-linear regression',xm_test,ym_test,ym_pred2)
```

```python
mb1 = dnn(bs,None)
mb1.fit(xb_train,yb_train,epochs = 20, validation_split = 0.2)

mb2 = dnn(bs,'relu')
mb2.fit(xb_train,yb_train,epochs = 20, validation_split = 0.2)
```

```python
yb_pred1 = mb1.predict(xb_test)
yb_pred2 = mb2.predict(xb_test)
```

```python
report('blobs linear regression',xb_test,yb_test,yb_pred1)
print("")
report('blobs non-linear regression',xb_test,yb_test,yb_pred2)
```

```python
def reportSVM(type,x_test,y_test,y_pred):
    y_pred1 = y_pred
    tp,fn,fp,tn = 0,0,0,0

    for i in range(len(y_test)):
        if  y_test[i] == 1 and y_pred1[i] == 1:
            tp += 1 
        if y_test[i] == 1 and y_pred1[i] == -1 :
            fn +=1
        if y_test[i] == -1 and y_pred1[i] == -1:
            tn += 1
        if y_test[i] == -1 and y_pred1[i] == 1:
            fp += 1

    print(type,":")
    print(f"tp = {tp} , fn = {fn}")
    print(f"fp = {fp} , tn = {tn}")
    
    
    plt.figure(figsize=(20,6))
    plt.subplot(1,2,1)
    plt.scatter(x_test[:,0],x_test[:,1],c=y_test)
    plt.title('true')
    
    plt.subplot(1,2,2)
    plt.scatter(x_test[:,0],x_test[:,1],c=y_pred1)
    plt.title('predicted')
    plt.suptitle(type)
    plt.show()
    
```

```python
reportSVM('SVM linear', xms_test, yms_test, yms_pred1)
print("")
reportSVM('SVM non-linear', xms_test, yms_test, yms_pred2)
```
